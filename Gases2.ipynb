### **a. Ley de Boyle**

```python
Data_Boyle = pd.read_csv("Boyle.csv")
df1 = pd.DataFrame(Data_Boyle)
df1
```

```python
mediana = df1["x"].median()
```

```python
df1['x1'] = 1 / df1['x']
df1
```

```python
plt.scatter(df1["x"], df1["y"])
plt.xlabel('Volumen(L)')
plt.ylabel('Presion(atm)')
plt.title('Ley de Boyle')
plt.show()
```

Para hacer lineal la gráfica reemplazamos el volumen por 1/volumen

```python
sns.regplot(x="x1",y="y", data = df1)
plt.xlabel('1 / Volumen(1/L)')
plt.ylabel('Presion(atm)')
plt.title("La ecuación de la recta es y = 30188x + 73.769")
plt.suptitle("Ley de Boyle")
plt.show()
```

### **d. Número de moleculas vs Velocidad**

```python
draw_histogram()
```

```python
df4 = pd.read_csv("Numero vs Velocidad.csv")
df4
```

```python
leyenda = {'a': '100 K', 'b': '300 K', 'c': '700 K'}

fig, ax = plt.subplots()
for nombre, grupo in df4.groupby('z'):
    etiqueta = leyenda.get(nombre, nombre)
    ax.plot(grupo['x'], grupo['y'], label=f'{etiqueta}')

plt.xlabel('Velocidad Molecular(m/s)')
plt.ylabel('Número de Moléculas')
plt.legend()
plt.title('# Moléculas vs Velocidad')
plt.grid(True)

plt.show()
```

Lo que significa el grafico es la probabilidad de las moléculas a tener una mayor velocidad dependiendo de la temperatura que se tenga sin variar; es decir, una temperatura constante, así mientras la temperatura sea mayor, hay mayor número de moléculas que tienen probabilidad de moverse más rápido.

## **2. Extraer datos químicos sobre gases disponibles en la web y hacer una exploración gráfica y estadística**

```python
url = 'https://www.ier.unam.mx/~ojs/pub/Liquid3/node9.html'
df = pd.read_html(url)

df [1]
```

```python
dfs = pd.read_html(url)
```

```python
df = dfs[1]

plt.figure(figsize=(10, 6))
plt.plot(df.iloc[1:, 1], df.iloc[1:, 2])
plt.xlabel("Temperatura Critica(°C)")
plt.ylabel('Presion critica(amt)')
plt.title('Puntos criticos de ciertos gases')
plt.show()
```

```python
df = dfs[1]

plt.figure(figsize=(10, 6))
plt.scatter(df.iloc[1:, 0], df.iloc[1:, 3])
plt.xlabel("Gas")
plt.ylabel('Punto de Ebullicion')
plt.title('Puntos de ebullicion de ciertos gases')
plt.show()
```
